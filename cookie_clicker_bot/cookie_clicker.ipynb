# Bot for Cookie Clicker

https://orteil.dashnet.org/cookieclicker/

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common import (
    ElementClickInterceptedException,
    ElementNotInteractableException,
    NoSuchElementException,
    StaleElementReferenceException
)
from time import sleep
```

```python
class CookieClicker:
    def __init__(self):
        self.SITE_LINK = "https://orteil.dashnet.org/cookieclicker/"
        self.PATH = r"C:\WebDrivers\chromedriver.exe"
        self.SITE_MAP = {
            "buttons": {
                "language": {
                    "portuguese": {"xpath", '//*[@id="langSelect-PT-BR"]'},
                    "english": {"xpath": '//*[@id="langSelect-EN"]'}
                },
                "cookie": {"xpath": '//*[@id="bigCookie"]'},
                "accept cookies": {"xpath": "/html/body/div[1]/div/a[1]"},
                "product": {"xpath": '//*[@id="product$$NUMBER$$"]'},
                "upgrade": {"xpath": '//*[@id="upgrade$$NUMBER$$"]'},
                "options": {
                    "xpath": '//*[@id="prefsButton"]/div',
                    "disable": {
                        "fancy": {"xpath": '//*[@id="fancyButton"]'},
                        "filters": {"xpath": '//*[@id="filtersButton"]'},
                        "numbers": {"xpath": '//*[@id="numbersButton"]'},
                        "particles": {"xpath": '//*[@id="particlesButton"]'},
                        "milk": {"xpath": '//*[@id="milkButton"]'}
                    },
                    "export save": {
                        "xpath": '/html/body/div/div[2]/div[18]/div[2]/div[4]/div[3]/div/div[4]/a[1]',
                        "text": {"xpath": '//*[@id="textareaPrompt"]'},
                        "close": {"xpath": '//*[@id="promptOption0"]'}
                    },
                    "import save": {
                        "xpath": '/html/body/div/div[2]/div[18]/div[2]/div[4]/div[3]/div/div[4]/a[2]',
                        "text": {"xpath": '//*[@id="textareaPrompt"]'},
                        "close": {"xpath": '//*[@id="promptOption0"]'}
                    },
                    "save": {"xpath": '//*[@id="menu"]/div[3]/div/div[3]/a'}
                }
            }
        }

        self.driver = webdriver.Chrome(service=Service(self.PATH))
        self.driver.maximize_window()

    def open_site(self):
        self.driver.get(self.SITE_LINK)
        sleep(10)

    def select_language(self):
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["language"]["english"]["xpath"]).click()
        sleep(5)

    def accept_cookies(self):
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["accept cookies"]["xpath"]).click()
        sleep(1)

    def settings(self):
        self.open_options()

        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["disable"]["fancy"]["xpath"]).click()
        sleep(0.1)
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["disable"]["filters"]["xpath"]).click()
        sleep(0.1)
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["disable"]["particles"]["xpath"]).click()
        sleep(0.1)
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["disable"]["milk"]["xpath"]).click()
        sleep(0.1)

        self.close_options()

    def click_cookie(self):
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["cookie"]["xpath"]).click()

    def get_product(self):
        for product in range(19):
            object_path = self.SITE_MAP["buttons"]["product"]["xpath"].replace("$$NUMBER$$", str(product))
            object_class = self.driver.find_element("xpath", object_path).get_attribute("class")

            if object_class == "product unlocked enabled":
                return str(product)
        return False

    def get_upgrade(self):
        for upgrade in range(5):
            object_path = self.SITE_MAP["buttons"]["upgrade"]["xpath"].replace("$$NUMBER$$", str(upgrade))
            object_class = self.driver.find_element("xpath", object_path).get_attribute("class")

            if object_class == "crate upgrade enabled":
                return str(upgrade)
        return False

    def buy_product(self):
        if self.get_product():
            object_path = self.SITE_MAP["buttons"]["product"]["xpath"].replace("$$NUMBER$$", self.get_product())
            self.driver.find_element("xpath", object_path).click()

    def buy_upgrade(self):
        if self.get_upgrade():
            object_path = self.SITE_MAP["buttons"]["upgrade"]["xpath"].replace("$$NUMBER$$", self.get_upgrade())
            self.driver.find_element("xpath", object_path).click()

    def open_options(self):
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["xpath"]).click()
        sleep(1)

    def close_options(self):
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["xpath"]).click()
        sleep(1)

    def export_save(self):
        self.open_options()

        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["save"]["xpath"]).click()

        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["export save"]["xpath"]).click()
        sleep(0.5)
        codigo = self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["export save"]["text"]["xpath"]).text
        with open("save.txt", "w") as f:
            f.write(codigo)
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["export save"]["close"]["xpath"]).click()

        self.close_options()

    def import_save(self):
        self.open_options()

        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["import save"]["xpath"]).click()
        sleep(0.5)
        with open("save.txt", "r") as f:
            codigo = f.read()
            self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["import save"]["text"]["xpath"]).send_keys(codigo)
        self.driver.find_element("xpath", self.SITE_MAP["buttons"]["options"]["import save"]["close"]["xpath"]).click()

        self.close_options()

    def exit(self):
        self.driver.quit()
```

```python
jogo = CookieClicker()
```

```python
try:
    jogo.open_site()
    jogo.select_language()
    jogo.accept_cookies()
    jogo.settings()
    # jogo.import_save()

    i = 0
    while True:
        if not i % 100 and i != 0:
            jogo.buy_product()
            sleep(0.1)
            jogo.buy_upgrade()

        jogo.click_cookie()
        i += 1
except (KeyboardInterrupt, KeyError, AttributeError,
        ElementClickInterceptedException, ElementNotInteractableException,
        NoSuchElementException, StaleElementReferenceException) as err:
    print(err)
    jogo.export_save()
    jogo.exit()
```
